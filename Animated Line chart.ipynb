```python
from sjvisualizer import DataHandler
from sjvisualizer import Canvas
from sjvisualizer import LineChart
from sjvisualizer import StaticText
```

```python
FPS = 60
DURATION = 0.25
BACKGROUND_COLOR = (0, 0, 0)
FONT_COLOR = (255,255,255)
colors = {
    "United States": (255,255,255)
}

events = {
    # "event_name": (start_date, end_date)
        "World War I": ["28/07/1914", "11/11/1918"],
        "The Great Depression": ["01/01/1929", "01/01/1939"],
        "World War II": ["01/01/1940", "31/12/1946"],
        "1973 Oil Crisis": ["01/01/1973", "31/12/1973"],
        "Latin American Debt Crisis": ["01/01/1980", "01/01/1981"],
        "Asian Financial Crisis": ["01/01/1997", "01/01/1998"],
        "Global Financial Crisis": ["30/06/2007", "31/12/2009"],
        "Corona Virus Pandemic": ["15/03/2020", "21/12/2021"],
}
```

```python
df=DataHandler.DataHandler(excel_file="US National Debt.xlsx", number_of_frames=FPS*DURATION*60).df
```

```python
canvas=Canvas.canvas(bg=BACKGROUND_COLOR)
```

```python
line=LineChart.line_chart(canvas=canvas.canvas, df=df, colors=colors, events=events, font_color=FONT_COLOR, unit="%")
canvas.add_sub_plot(line)
```

```python
canvas.add_title("National debt of the United States", color=(255,255,255))
canvas.add_sub_title("As Percentage of GDP", color=(255,255,255))
```

```python
footnote=StaticText.static_text(canvas=canvas.canvas, text="Data Source: IMF and Economic Research Federal Reserve Bank of St. Louis", width=0, height=25, x_pos=50, y_pos=1000, anchor="w", font_color=(255,255,255))
canvas.add_sub_plot(footnote)
```

```python
canvas.add_time(time_indicator="year", df=df)
```

```python
canvas.play(fps=FPS)
```
